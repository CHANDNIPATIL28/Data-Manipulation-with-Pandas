```python
import pandas as pd
```

```python
# Data for all of the locations of Petal Power is in the file inventory.csv
df = pd.read_csv('inventory.csv')
df.head()
```

```python
# The first 10 rows represent data from your Staten Island location
staten_island = df.iloc[0:10]
staten_island
```

```python
# A customer just emailed you asking what products are sold at your Staten Island location
product_request = staten_island['product_description']

# Another customer emails to ask what types of seeds are sold at the Brooklyn location        
seeds_request = df[(df.location == 'Brooklyn') & (df.product_type == 'seeds')]
```

```python
# Adding a column to inventory called in_stock which is True if quantity is greater than 0 and False if quantity equals 0
df['in_stock'] = df.quantity.apply(lambda x: True if x > 0 else False)
df.head(10)
```

```python
# Petal Power wants to know how valuable their current inventory is.
df['total_value'] = df.apply(lambda row: row['price']*row['quantity'],axis=1)
df.head(10)
```

```python
# The Marketing department wants a complete description of each product for their catalog.
combine_lambda = lambda row: \
    '{} - {}'.format(row.product_type,
                     row.product_description)
df['full description'] = df.apply(combine_lambda,axis=1)
df.head(10)
```
